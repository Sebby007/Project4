1. The Necessary libraries are imported

```python
# Import the required libraries
import pandas as pd
import plotly.express as px
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np
```

2. Load the data set from the downloads folder and show the data to get a first glance of what it contains

```python
# Load the data
Athletes = pd.read_csv("C:/Users/sebas/Downloads/Athletes.csv")
```

```python
# Dropping unnecessary columns
Athletes.drop(['Wikipedia Page', 'dbpedia Page', 'Image', 'Description'], axis=1, inplace=True)
```

```python
display(Athletes)
```

3. The last 4 columns contained information like links to wikipedia about each individual player and their corresponding wikipedia picture. This information is not very useful or relevant for statistical analysis, so I made de decision to get rid of it entirely. 

```python
print(Athletes.info())
```

4. Using the info method above to get a better insight of the data: number of rows and columns, data types and missing values

```python
# Remove dollar signs and commas
Athletes['Endorsements'] = Athletes['Endorsements'].replace({'/$': '', ',': ''}, regex=True)
Athletes['Total Pay'] = Athletes['Total Pay'].replace({'/$': '', ',': ''}, regex=True)

# Convert to numeric with coercion to handle errors
Athletes['Endorsements'] = pd.to_numeric(Athletes['Endorsements'], errors='coerce')
Athletes['Total Pay'] = pd.to_numeric(Athletes['Total Pay'], errors='coerce')

# Check for any conversion issues
print(Athletes['Endorsements'].isnull().sum()) 
print(Athletes['Total Pay'].isnull().sum()) 
```

5. Dollar signs are removed and the data type of the columns Total Pay and Endorsements is changed to Numeric to facilitate arithmethic operations if needed. 

```python
display(Athletes)
```

```python
print(Athletes.describe())
```

6. Usinng the describe() method to get insights on numeric data such as the height of the atletes, thit total pay and endorsements. 

```python
# Enable inline plotting
%matplotlib inline

# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(Athletes['Sport'], Athletes['Total Pay'])

# Rotate x-axis labels for better readability
plt.xticks(rotation=30)

# Add labels and title
plt.xlabel('Sport')
plt.ylabel('Total Pay')
plt.title('Total Pay of Athletes by Sport')

# Display the plot
plt.show()
```

7. Creating a bar plot to visualize the distribution of Sports and payments per year, the top 3 highest paid sports are Boxing, soccer and basketball acording to this graphic. 

```python
# Group the data by Sport and Gender, then count
data = Athletes.groupby(['Sport', 'Gender']).size().unstack()

# Plot
plt.figure(figsize=(10, 6))
data.plot(kind='bar', stacked=True, ax=plt.gca())

# Labels and title
plt.xlabel('Sport')
plt.ylabel('Count of Athletes')
plt.title('Gender Distribution in Sports')
plt.xticks(rotation=30)

# Display the plot
plt.show()
```

8. Finally a distribution of data between sports and gender is plotted. It is observed that male athletes dominate in almost all sports with the exception of tennis.
